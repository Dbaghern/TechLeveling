# Applying Randomness to Probability

To answer questions about proability, we usually calculate exact probabilites. However, sometimes probability isn't always easy to calculate. We can use the properties of random processes to approximate an exact caluclation.

## Monte Carlo Simulation

Monte Carlo is a family of methods designed to do this. While just looking at one run of an experiment might not be helpful, if we randomly run an experiment thousands of times, we can aggregate results to get an estimate of the overall probability. 

Taking the results from above, if we normalize over the sample size, we get the relative probability of runs of length n from a random coin flip. The larger we make our simulation, the closer the approximation will be to the actual value.

```python
plt.figure(figsize=(10,6))

plt.hist(runs1,bins=range(0, 15, 1), histtype='stepfilled', normed = True, color='b', label='Computer')

plt.title("Probability of Runs of Length N")
plt.xlabel("Length of Run")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

# The Birthday Problem

What is the probability of a at least two people in a group of n people sharing a birthday?

Given three people, you can solve the problem by considering possible outcomes and adding the respective probabilities together. First we determine:
- the probability of person 1 and person 2 sharing a birthday
- the probability of person 1 and person 3 sharing a birthday
- the probability of person 2 and person 3 sharing a birthday
- the probability of person 1, person 2, and person 3 sharing a birthday

Then, we just add those individual probabilites together to get the final probability.

As the size of the group grows, the number of computations grows much faster than the size of the group!


### Disjoint Events

An event is a collection of possible outcomes of some sample space. In a coin flip, getting heads is an event, getting tails is an event, and getting a heads or tails is an event. 
Two events are disjoint if they share no common occurences. So P(H), getting heads, is disjoint from P(T), getting tails, but P(H) is not disjoint from P(H or T), getting heads or tails.

The disjoint events that are useful when they cover all possible outcomes of a sample space. In our coin flip example, P(H) and P(T) are disjoint, and when considered together, cover all possible outcomes.

When we have disjoint events that cover the whole space, no matter what the individual probabilities are, together they add to 100%. From this, we derive the formula P(H) + P(T) = 1. Or, for disjoint sets E1 and E2 covering the whole space, P(E1) + P(E2) = 1.


### Back to Birthdays

We can frame any specific event in the context of disjoint events by considering that for any event, either it occurs or it doesn't occur. Thus for any event, P(Event) + P(Not Event) = 1. 
 
If we are interestd in the probability of at least two people sharing a birthday, the opposite would be that no one could share a birthday. We can now say that P(at least 2 out of n people sharing a birthday) + P(0 out of n people sharing a birtday) = 1


### Calculating the Opposite

How can we calculate the opposite probability? 

We have to calculate the individual constraints, then multiply the probabilities together. So for three poeple, I need to calculate the respective probability for each person, then multiply them together. 

Recall, a probability is the number of ways to achieve the event out of the number of all possible outcomes. So for a coin toss, P(H) = 1/2, and P(H or T) = 2/2. 

For the first person, there is no constraints, we can pick any day out of the year and we won't have a matching birthday. We can choose any of the 365 days (let us ignore leap years). Thus we have a 365/365 chance of getting a non matching birthday for the first person.

For the second person, they cannot have the same day. The day itself is arbitrary, but they only have 364 days to choose from, giving us a 364/365 probability for them not having a matching birthday.
 
For the third person, they then have 363 days to choose, and if we wanted to continue in this manner, we would keep subtracting one for each successive person.

To calculate the combined probability, we just multiply the individual probabilites together.
- The probability of 1 person not sharing a birthday is 365/365. 
- The probability of two people not sharing a birthday is 365/365 * 364/365.
- The probability of three people not sharing a birthday is 365/365 * 364/365 * 363/365.
- And so on.


Now we can use the formula to find the opposite! Using algebra, we see that for three people:

- P(at least 2 out of 3 people sharing a birthday) + P(0 out of 3 people sharing a birtday) = 1
- P(at least 2 out of 3 people sharing a birthday) = 1 - P(0 out of 3 people sharing a birtday) 
- P(at least 2 out of 3 people sharing a birthday) = 1 - 365/365 * 364/365 * 363/365 

```python
from __future__ import division #floating point division for python 2.7

# n = number of people in the group
def birthday_probability(n):
    prob = 1      # Initializing to 1 for loop multiplication
    
    for i in range(0,n):
        prob = prob * (365-i)/365
    
    # Final probability of n people sharing a birthday
    return (1 - prob)



n = 8
print "Probability of at least two of " + str(n) + " people sharing a birthday: " 
print birthday_probability(n)
```

### Function Exploration

Now that we know how to calculate an individual probability, let us look at the underlying function to answer some questions.

```python
import pandas as pd
import numpy as np

# creating a data frame for graphing
df = pd.DataFrame({ 'People' : range(365)})

## Applying our function to each row in the data frame to get the respective probability
df['Probability'] = df['People'].apply(birthday_probability)

#Visual inspection
df.head(10)
```

### Monte Carlo Simulation

In this example, the probability was easy to calcuate once we rephrased the problem. We can also run a Monte Carlo simulation to approximate the probability like we did above with coin flips. To do this, lets generate a lot of random samples of n people.

```python
import random

n = 8
numtrials = 10000
# total counts of matches 
match = 0


for i in range(numtrials):
    #generating a random sample of size n
    sequence =[random.randrange(1,366,1) for _ in range (n)]
    
    #if there are duplicates in the array, the set of distinct elements will be smaller
    if len(sequence) != len(set(sequence)):
        match += 1

        
print "Number of matches: " + str(match)
print "Relative probability: " + str(match/numtrials)
```

### Calculating Specific Probabilities
 
This probability is valid based on the assumption that all birthdays have an equal probability assigned. Is that really the case?

Finding a public dataset of birthdays is really difficult! The best dataset I could find was a sample dataset within Google BigQuery. To pull the counts of birthdays, use this SQL query:

    SELECT MONTH, DAY, COUNT(*) AS FREQUENCY
    FROM [publicdata:samples.natality]
    WHERE DAY >= 1 AND DAY <= 31
    GROUP BY MONTH, DAY
    ORDER BY MONTH, DAY
    
The resulting data was saved in the birthdays.csv file. Let us use this to calculate the relative probabilities from the population.

```python
#loading csv into data frame
birthdays = pd.read_csv("birthdays.csv")

print 'Structure of csv file:\n'
print birthdays.head()
print '\nRelative Birthday Probabilities:\n'
print birthdays['frequency']/sum(birthdays['frequency'])
```

```python
#Creating a new column to use as x axis values
birthdays['date'] = birthdays['month'].astype(str) + '-' + birthdays['day'].astype(str)
#Caluclating probabilities for y axis values
birthdays['probabilities'] = birthdays['frequency']/sum(birthdays['frequency'])

birthdays.plot(x='date', 
        y='probabilities', 
        legend=False, 
        title="Probability of Birthdays",
        xticks = range(0, 365, 20),
        rot = 30, #rotation of axis labels
        fontsize = 12,
        figsize = (10,6)) #picture dimensions

```

### Adjusting the Simulation

We can see that while in general the relative probabilities are the same, there are a few variations over the year. Does this affect our initial calculations?


```python
n = 8

match = 0
numtrials = 100000

for i in range(numtrials):
    sequence = np.random.choice(len(birthdays['date']), n, p=birthdays['probabilities'])
    sequence =[random.randrange(1,366,1) for _ in range (n)]
    if len(sequence) != len(set(sequence)):
        match += 1
        
print "Number of matches: " + str(match)
print "Relative probability: " + str(match/numtrials)
```

# Monty Hall

In the game show Let's Make a Deal, you play by picking one of three doors. Whatever is behind the door, you get to keep. Behind one door is a fancy car, behind the other two are goats.

After choosing a door, the host decides to throw a curveball: he opens one of the other two doors to reveal a goat. Now that you know what's behind one of the doors, he gives you a choice: you can open the door you originally chose, or you can switch to the other unopened door. Which do you choose?


### Precise Calculation
To calculate the probability by hand, we need to make use of disjoint events like we did before.

For your initial door, you have a 1/3 chance of picking the winning door. Once the host opens a door, what changes?

If the host opened the door with the car behind it, the game is over, you lose! That doesn't make for compelling tv, so he has to choose the door with a goat. Because he did not randomly choose a door, that door is still in play in the calculations. 

- P(other door having a car) = 1 - P(other door not having a car)
- P(other door having a car) = 1 - P(first door or opened door having a car)
- P(other door having a car) = 1 - P(first door not having a car) - P(opened door having a car)
- P(other door having a car) = 1 - 1/3 - 0
- P(other door having a car) = 2/3

You gain an additional 33% by switching to the other opened door, simply because the event is not random. 
  
  
### Monte Carlo Calculation

We can also derive this 33% gain by using Monte Carlo. This simulation is a little more complicated, but you just need to define the experiment properly to get the correct results.

Begin by making a basic simulation of one event.

```python
#initializing doors
doors = np.array([0,0,0])
#randomly pick a door and assign it to a door
car = np.random.choice(3, 1) 
doors[car] = 1

print "Prizes behind doors: " + str(doors)

#randomly pick a door 
contestant_choice = np.random.choice(3, 1)
print "Contestant choice: " + str(contestant_choice)

# host has to choose one of the other two doors
host_options = np.array(range(3))
mask = np.in1d(host_options, contestant_choice, invert=True)
host_options = host_options[mask]
print "Door options for host: " + str(host_options)

#if the prize is in the options, choose the other door
if(1 in doors[host_options]):
    open_door = doors[host_options] != 1
    print 'Prize is in host options, has to pick other door'
    open_door = host_options[open_door]
    print open_door 
else:
    #otherwise randomly choose a door
    print 'Prize is not in host options, random choice'
    open_door = np.random.choice(host_options, 1) 

print "Opened door: " + str(open_door)

new_choice = host_options[host_options != open_door]
print "Contestant's new choice: " + str(new_choice)

# Should the contestant switch? 
win_stayed = contestant_choice == car
win_switched = new_choice == car

print "Win if he switches? " + str(win_switched)
print "Win if he stays? " + str(win_stayed)
```

### Aggregation

Now that we have a simulation working, we just repeat it thousands of times and pool the results.

```python
total_stays = 0
total_switches = 0
ntrials = 10000

for trial in range(ntrials):
    #initializing doors
    doors = np.array([0,0,0])
    #randomly pick a door and assign it to a door
    car = np.random.choice(3, 1) 
    doors[car] = 1

    #randomly pick a door 
    contestant_choice = np.random.choice(3, 1)

    host_options = np.array(range(3))
    mask = np.in1d(host_options, contestant_choice, invert=True)
    host_options = host_options[mask]

    #if the prize is in the options, choose the other door
    if(1 in doors[host_options]):
        open_door = doors[host_options] != 1
        open_door = host_options[open_door]
    else:
        #otherwise randomly choose a door
        open_door = np.random.choice(host_options, 1) 


    new_choice = host_options[host_options != open_door]

    # Should the contestant switch? 
    win_stayed = contestant_choice == car
    win_switched = new_choice == car

    total_stays = total_stays + int(win_stayed)
    total_switches = total_switches + int(win_switched)
    
print "Probability of winning if he keeps first choice: " + str(total_stays /ntrials)
print "Probability of winning if he switches: " + str(total_switches / ntrials)
print "Overall gain by switching: " + str((total_switches / ntrials) - (total_stays /ntrials))
```
